```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import plotly.graph_objects as go
import plotly.express as px
```

```python
# importing the dataset

titanic_DS = pd. read_csv("Titanic Data.csv")
titanic_DS
```

```python
#checking for missing values
print(titanic_DS.isnull().sum())
```

So Age, Cabin and Embarked have missing values and thus carry out imputation

```python
titanic_DS['Age'].fillna((titanic_DS['Age'].median()), inplace=True)
titanic_DS['Cabin'] = titanic_DS['Cabin'].fillna((0))
titanic_DS['Embarked'] = titanic_DS['Embarked'].fillna((0))
titanic_DS.reset_index()
```

```python
print(titanic_DS.isnull().sum())
```

Answer 1: Distribution of age using a histogram

```python
fig = px.histogram(titanic_DS, x="Age", nbins=20, title='A histogram showing the distribition of Age')

fig.show()
```

Answer 2: Distribution of age based on gender using a histogram

```python
fig = px.histogram(titanic_DS, x="Age", y= 'Sex', nbins = 20, title = 'A histogram showing the distribution of age based on gender' )

fig.show()
```

Answer 3 : A correlation matrix of the data

```python
correlation_matrix = titanic_DS.corr()

print(correlation_matrix)
```

```python
fig = px.imshow(correlation_matrix,color_continuous_scale='rdylbu', title = 'A graph showing the correlation matrix of the data')

fig.show()
```

Answer 4: A pivot table of the Fare vs Pclass

```python
pivot_table = titanic_DS.pivot_table(values  = 'Fare', columns = 'Pclass')

print(pivot_table)
```

```python
fig = px.box(titanic_DS, x="Pclass", y="Fare", title='A box plot showing the distribution of Pclass vs Fare')

fig.show()
```

Answer 5 : A stacked bar graph showing those who survived and did not survive, based on gender

```python

fig = px.bar( titanic_DS, x = 'Sex' ,  color = 'Survived', barmode='stack', title = 'A stacked bar graph showing the survival based on gender')
fig.show()
```

Answer 6: Using a scatter plot, plot the ages to the fare paid by the each passenger based on their
gender.

```python
fig = px.scatter(titanic_DS, x="Fare", y="Age", color="Sex",
                title="A scatter plot showing the fare paid by each individual based on gender and age")

# show plot 
fig.show()
```

Answer 7: Plot a bubble plot of the ages to the fare paid by each passenger categorizing whether
they survived or not.

```python
fig = px.scatter(titanic_DS, x="Fare", y="Age",
	         size="Pclass", color="Survived",
                 hover_name="Name", size_max=50, title='A bubble plot of ages to fare distinguished by survival')
fig.show()
```
